```python
from bs4 import BeautifulSoup

# Supongamos que ya tienes la página web en la variable 'html'
soup = BeautifulSoup(html, 'html.parser')

# Crear un diccionario para almacenar los poemas
poemas_diccionario = {}

# Obtener todos los elementos con etiqueta 'b' que representan los títulos de los poemas
titulos_poemas = soup.find_all('b')

# Recorrer cada título y obtener los párrafos y oraciones correspondientes
for titulo_poema in titulos_poemas:
    # Obtener el título del poema
    titulo = titulo_poema.text

    # Obtener los párrafos del poema
    parrafos = []
    siguiente_elemento = titulo_poema.next_sibling  # Primer párrafo

    while siguiente_elemento:
        # Verificar si es un salto de línea o una etiqueta de párrafo
        if siguiente_elemento.name == 'br':
            siguiente_elemento = siguiente_elemento.next_sibling
            continue

        # Obtener las oraciones dentro del párrafo
        oraciones = siguiente_elemento.strip().split('\n')

        # Agregar el párrafo a la lista de párrafos
        parrafos.append(oraciones)

        siguiente_elemento = siguiente_elemento.next_sibling  # Siguiente párrafo o título

    # Agregar el título, párrafos y oraciones al diccionario
    poemas_diccionario[titulo] = parrafos

# Imprimir el diccionario con la estructura de poemas
for titulo, poemas in poemas_diccionario.items():
    print("Título:", titulo)
    for i, parrafo in enumerate(poemas):
        print("Párrafo", i+1)
        for j, oracion in enumerate(parrafo):
            print("Oración", j+1, ":", oracion)
        print()
```

```python
from bs4 import BeautifulSoup

# Supongamos que ya tienes la página web en la variable 'html'
soup = BeautifulSoup(htmlR, 'html.parser')

# Crear un diccionario para almacenar los poemas
poemas_diccionario = {}

# Obtener todos los elementos con etiqueta 'b' que representan los títulos de los poemas
titulos_poemas = soup.find_all('b')

# Recorrer cada título y obtener los párrafos y oraciones correspondientes
for titulo_poema in titulos_poemas:
    # Obtener el título del poema
    titulo = titulo_poema.text

    # Obtener los párrafos del poema
    parrafos = []
    siguiente_elemento = titulo_poema.next_sibling  # Primer párrafo

    while siguiente_elemento:
        # Verificar si es un salto de línea o una etiqueta de párrafo
        if siguiente_elemento.name == 'br':
            # Verificar si es un salto de línea adicional para separar los párrafos
            if siguiente_elemento.next_sibling and siguiente_elemento.next_sibling.name == 'br':
                # Agregar el párrafo actual a la lista de párrafos
                poemas_diccionario[titulo].append(parrafos)
                # Inicializar una nueva lista para el siguiente párrafo
                parrafos = []
            siguiente_elemento = siguiente_elemento.next_sibling
            continue

        # Obtener las oraciones dentro del párrafo
        oraciones = siguiente_elemento.text.strip().split('\n')

        # Agregar el párrafo a la lista de párrafos
        parrafos.append(oraciones)

        siguiente_elemento = siguiente_elemento.next_sibling  # Siguiente párrafo o título

    # Agregar el título y la lista de párrafos al diccionario
    poemas_diccionario[titulo] = parrafos
print(poemas_diccionario.items())
```
