```python
import datetime
from datetime import datetime
from datetime import timedelta

print("-------------------BIENVENIDO A SU LIBRETA DE AHORRO---------------------")
print("-------------------------------------------------------------------------")
print("Ocion 1 ------> Realizar Depócito")
print("Ocion 2 ------> Realizar Retiro")
print("Ocion 3 ------> Mostrar Libreta de Ahorros")
print("Ocion 4 ------> Salir")
print("-------------------------------------------------------------------------")
Opcion = int(input("Ingrese Nro de Opción:  "))
Libreta = []
Monto = 0
FechaAnt = datetime(2016,1,1)
Tasa = 0
while (Opcion != 4):
  if (Opcion == 1):
    Fecha = datetime.strptime(input("Ingrese Fecha de Depocito:  "), "%d-%m-%Y")
    Depocito = int(input("Ingrese el Monto del Depócito (s/.): "))
    Interes = ((Tasa/100)/30)*((Fecha-FechaAnt).days)*Monto
    Tasa = int(input("Por favor Ingrese la tasa de interes mensual (%):  "))
    Monto = Monto + Depocito + Interes
    FechaAnt = Fecha
    Libreta.append([datetime.strftime(Fecha,"%d-%m-%Y"), Depocito, 0, Tasa, round(Interes,2), round(Monto,2)])
  elif(Opcion == 2):
    Fecha = datetime.strptime(input("Ingrese Fecha de Retiro:  "), "%d-%m-%Y")
    Retiro = int(input("Ingrese el Monto del Retiro (s/.): "))
    Interes = ((Tasa/100)/30)*((Fecha-FechaAnt).days)*Monto
    Tasa = int(input("Por favor Ingrese la tasa de interes mensual de su Banco (%):  "))
    Monto = Monto - Retiro + Interes
    FechaAnt = Fecha
    Libreta.append([datetime.strftime(Fecha,"%d-%m-%Y"), 0, Retiro, Tasa, round(Interes,2), round(Monto,2)])
  elif(Opcion == 3):
    print("|     FECHA     |  |   Depócito   |  |   Retiro   |  |   Tasa   |  |   Interes   |  |   Monto   |")
    for k in Libreta:
      print("    "+str(k[0])+"          "+str(k[1]).ljust(17," ")+str(k[2]).ljust(17," ")+str(k[3])+"%".ljust(14," ")+str(k[4]).ljust(15," ")+str(k[5]))
      print("----------------------------------------------------------------------------------")

  print("-------------------------------------------------------------------------")
  print("Ocion 1 ------> Realizar Depócito")
  print("Ocion 2 ------> Realizar Retiro")
  print("Ocion 3 ------> Mostrar Libreta de Ahorros")
  print("Ocion 4 ------> Salir")
  print("-------------------------------------------------------------------------")
  Opcion = int(input("Ingrese Nro de Opción:  "))

print(" ")
print(" ")
print("GRACIAS POR USAR NUESTRA APLICACION")


```
