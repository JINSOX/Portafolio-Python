```python
from fim import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
from pymining import seqmining
from paretoset import paretoset
```

```python
#converting dataframe into list of lists
dataframe_list=[]

# values of range is according to the number of rows and columns
for i in range(1,7501):
    dataframe_list.append([str(df.values[i,j]) for j in range(0,20)])
```

```python
dataframe_list
```

```python
dataframe_list_limpios = [[item for item in transaccion if item != 'nan'] for transaccion in dataframe_list]
```

```python
dataframe_list_limpios
```

Aplicar el algoritmo apriori y obtener reglas de asociación
Algoritmo APRIORI

```python
# Creamos un data frame con pandas
te = TransactionEncoder()
te_ary = te.fit(dataframe_list_limpios).transform(dataframe_list_limpios)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
# Aplicamos el algoritmo Apriori para obtener conjuntos de itmes frecuentes
frequent_itemsets = apriori(df, min_support=0.015, use_colnames=True )
frequent_itemsets
```

### Generación de reglas de asociación

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.0)
rules
```

### Filtrar las mejores 3 reglas

```python
# Filtrar las mejores 10 reglas según criterios definidos
top_3_rules = rules.sort_values(by=['support', 'confidence'], ascending=False).head(3)
top_3_rules
```

```python
# Seleccionar tres reglas para comentar
selected_rules = top_3_rules.head(3)

# Comentar la calidad de las reglas
for i, rule in enumerate(selected_rules.itertuples(), 1):
    antecedent = set(rule.antecedents)
    consequent = set(rule.consequents)

    print(f"Regla {i}:")
    print(f"Antecedente: {antecedent}")
    print(f"Consecuente: {consequent}")
    print(f"Support: {rule.support:.3f}")
    print(f"Confidence: {rule.confidence:.3f}")
    print(f"Lift: {rule.lift:.3f}")
    print("\n")
```

##Mining itemsets

```python
#todos los itemsets frecuentes con al menos 30% de soporte, y 1 atractivo como mínimo
r = fpgrowth(dataframe_list_limpios, supp=3, zmin=2)
r
```

```python
#función para obtener todos los itemsets frecuentes de un df
def all_itemsets(trans_, supp_=1):
  r = fpgrowth(trans_, supp=supp_, report='aS')
  df_items = pd.DataFrame(r)
  df_items.columns = ['Itemset', 'Freq', 'Freq(%)']
  df_items.sort_values(by='Freq', ascending=False, inplace=True)
  df_items['Size'] = [len(x) for x in df_items['Itemset'].values]
  df_items['Itemset'] = [str(sorted(x)) for x in df_items['Itemset'].values]
  return df_items

all_itemsets(dataframe_list_limpios, 2)
```

```python
mask = paretoset(dataframe_list_limpios[['Freq', 'Size']], sense=['max', 'max'])
sky_itemsets = dataframe_list_limpios[mask]
print(len(sky_itemsets))
sky_itemsets
```
